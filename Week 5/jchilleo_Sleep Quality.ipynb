```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sleep = pd.read_csv(r'C:\Users\Joseph\OneDrive\Desktop\Data Science\MSDS670-Data-Visualization\Week 5\Sleep_health_and_lifestyle_dataset.csv', index_col=0)
sleep.sample(10)
```

```python
sleep.info()
```

```python
occ_sleep = sleep.groupby('Occupation')['Quality of Sleep'].mean()
occ_sleep
```

```python
# reset index to convert the grouped result to a DataFrame
occ_sleep = occ_sleep.reset_index()
```

```python
# change Sales Representative to Sales Rep
occ_sleep['Occupation'] = occ_sleep['Occupation'].replace("Sales Representative", "Sales Rep")
```

```python
sns.barplot(data=occ_sleep, x='Occupation', y='Quality of Sleep',color='teal')
plt.xticks(rotation=80)
plt.title('Average Quality of Sleep by Occupation')
plt.ylim(1,10)
```

```python
sns.boxplot(data=sleep, x='Gender', y='Quality of Sleep')
plt.ylim(1,10)
plt.title('Sleep Quality by Gender')
```

```python
sns.regplot(data=sleep, x='Sleep Duration', y='Quality of Sleep',color='red',line_kws={'color':'black'})
plt.xlabel('Sleep Duration (Hours)')
plt.ylabel('Quality of Sleep')
plt.title('Sleep Quality by Duration')
plt.ylim(1,10)
```
